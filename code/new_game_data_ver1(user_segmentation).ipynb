```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.graph_objects as go
from datetime import datetime
%matplotlib inline
```

```python
df = pd.read_csv("project_groupB_data_240229.csv")
df.head()
```

```python
df.info()
```

```python
def convert_to_date_with_year(date_string):
    return datetime.strptime(f'{date_string}-{2024}', '%d-%b-%Y')
```

```python
df['first_login_date'] = df['first_login_date'].apply(convert_to_date_with_year)
```

```python
df.head()
```

```python
plt.hist(df['playtime'], bins=50, color='skyblue', edgecolor='black')  # bins   
plt.title('Playtime Histogram')
plt.xlabel('Playtime')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.hist(df['level'], bins=50, color='skyblue', edgecolor='black')  # bins   
plt.title('level Histogram')
plt.xlabel('level')
plt.ylabel('Frequency')
plt.show()
```

```python
#    
```

```python
df['t_score']= pd.cut(df['playtime'], bins=5, labels=['1', '2', '3', '4', '5'])
```

```python
df['l_score']= pd.cut(df['level'], bins=5, labels=['1', '2', '3', '4', '5'])
```

```python
df.head()
```

```python
df['total_score'] = df['t_score'].astype(str) + df['l_score'].astype(str)
```

```python
df.head()
```

```python
df['total_score_100'] = (0.5 * df['t_score'].astype(int) + 0.5 * df['l_score'].astype(int)) * 100 / 5
```

```python
df.head()
```

```python
df['user_tier']= pd.qcut(df['total_score_100'], q=5, labels=['Beginner', 'Bronze', 'Silver', 'Gold', 'Diamond'])
```

```python
df.head()
```
